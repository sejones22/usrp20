# Periodogram

```python
import matplotlib.pyplot as plt
from lightkurve import search_lightcurvefile
lc = search_lightcurvefile('G 227-22', mission = "TESS").download_all().PDCSAP_FLUX.stitch();
# Starname
#starname='G 227-22'
# Download all lightcurves
#lc = search_lightcurvefile(starname, mission="TESS").download_all()
```

```python
lc.remove_outliers().plot(); #; removes output
```

```python
pg = lc.remove_outliers().remove_nans().to_periodogram(oversample_factor=1)
pg.plot(view='period', scale='log');
```

```python
pg.period_at_max_power
```

```python
lc.remove_outliers().fold(pg.period_at_max_power, t0=1).scatter();
```
